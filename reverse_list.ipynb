## [::-1]

```python
my_list=[1,2,3,4]
reversed_list =my_list[::-1]
print(reversed_list)
print(my_list)
```

```python
list =["Mango", "Apple", "Cherry"]
reversed_list = list[::-1]
print(list)
print(reversed_list)
```
