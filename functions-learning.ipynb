## Write a function that returns the largest of two numbers.

```python
def largest_number(numbers):
    max_num = numbers[0]  # start with the first number
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num

print(largest_number([10, 50]))
```

## Write a function that counts how many times a letter appears in a word.


```python
def count_letter(word, letter):
    count = 0
    for i in word:
        if i == letter:
            count += 1
    return count
print(count_letter("Hello", "l"))
```

## Write a function that returns the grade of a student based on a score:
## 80–100 = A
## 60–79 = B
## 40–59 = C
## Below 40 = D

```python
def student_grade(score):
    if score >= 80 :
        print("grade: A")
    elif  score >= 60 and score<=79:
        print("grade: B")
    elif  score >= 40 and score<=59:
        print("grade: C")
    elif  score < 40 :
        print("grade: D")
    else:
        print("No mark")
    
student_grade(70)
    
```

## Write a function that finds the smallest number in a list.

```python
def smallest_number(numbers):
    min_num = numbers[0]
    for num in numbers:
        if num < min_num:
            min_num = num
        return min_num
smallest_number([10,20])
        
```

## Write a function that counts how many vowels are in a word.

```python
def count_vowels(word):
    vowels = "a,e,i,o,u,A,E,I,O,U"
    count = 0
    for char in word:
        if char in vowels:
            count += 1
    return count
count_vowels("Education")
        
```

## Write a function that removes all vowels from a string.

```python
def remove_vowels(word):
    vowels = "aeiouAEIOU"
    result = ""
    for item in word:
        if item not in vowels:
            result += item
    return result
print(remove_vowels("Game"))

    
```

## Write a function that returns the sum of all even numbers in a list.

```python
def sum_even(numbers):
    count = 0
    for num in numbers:
        if num % 2 == 0:
            count += num
        return count
sum_even([10,20,21])
```

## Challenge 1: FizzBuzz

#### Write a function that prints numbers from 1 to 100. For multiples of 3, print "Fizz" instead of 
#### the number. For multiples of 5, print "Buzz". For multiples of both 3 and 5, print "FizzBuzz". 

```python
def print_numbers():
    for num in range(1,100):
        if num % 3 == 0 and num % 5 == 0:
            print("FizzBuzz")  
        elif num % 3 == 0:
            print("Fizz")
        elif num % 5 == 0:
            print("Buzz")  
        else:
            print(num)
print_numbers()
```
