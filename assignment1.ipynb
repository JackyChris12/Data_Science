## Challenge 5: String Manipulation 

```python
# Given a string, print: 
# - Its length 
# - All uppercase 
# - All lowercase 
# - First 3 characters 
# - Last 3 characters 
# - Reversed string 
text = "Python Programming" 
print(len(text))
uppercase_letters = ''.join([char for char in text if char.isupper()])
print(uppercase_letters)

lowercase_letters = ''.join([char for char in text if char.lower()])
print(lowercase_letters)

print(text[0:3])
print(text[15:])
print(text[::-1])
```

## Challenge 9: Leap Year Checker 


```python
# Check if a year is a leap year 
# Rules: divisible by 4, but not by 100, unless also divisible by 400 
year = 2024
# Print "Leap year" or "Not a leap year" 
if year % 4 == 0:
    print("Leap year")
else:
    print("Not Leap year")
```
