```python
class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def bark(self):
        print("woof")

    def birthday(self):
        self.age += 1

d =Dog("Lucky", 2)
d.bark()
d.birthday()
print(d.age)
        

```

```python
class Student:
    def __init__(self,name, grade):
        self.name = name
        self.grade = grade

    def student_info(self):
        print(f"My name is {self.name} and I have a grade of {self.grade}")

s =Student("Jackline","A")
s.student_info()
```

```python
class Teacher:
    def __init__(self,name,subject,age,gender):
        self.name = name
        self.subject = subject
        self.age = age
        self.gender = gender

    def teacher_info(self):
        return f"My name is teacher {self.name}, a {self.gender} teacher of age {self.age} and I teach my favourite subject {self.subject}"

t =Teacher("Mercy","Python", 20,"Female")
t.teacher_info()
        
```

```python
class Hospital:
    def __init__(self,doctor,patient,bed):
        self.doctor = doctor
        self.patient = patient
        self.bed = bed

    def record(self):
        return f"In ward 2 {self.doctor} is taking care of {self.patient} in bed {self.bed}"

h =Hospital("James", "Tom", 4)
h.record()
```
