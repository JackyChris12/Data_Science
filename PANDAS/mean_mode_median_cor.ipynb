```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {'A': [10, 20, 30, 40],
        'B': [5, 15, 25, 35],
        'C': [1, 2, 4, 4]}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
## mean
mean_a = df['A'].mean()
mean_b = df['B'].mean()
mean_c = df['C'].mean()
print(mean_a)
print(mean_b)
print(mean_c)
```

```python
## Median
median_C= df["C"].mode()
median_C
```
