## Functions- block of code that performs specific task and can be reused


```python
def greet(name):
    return f"Hello, {name}!"
    
print(greet("Alice"))
```

## Function parameters- We have positional and Keyword arguments

```python
def info(name, age):
    print(f"{name} is {age} years old")

info("Bob", 30) ## positional argument
info(age = 25, name = "Jackline") # keyword agrument
```

## Default parameters- provides default values for parameters

```python
def greet(name="Guest"):
    print(f"Hello,{name}!")

greet()
```

## Variable length Arguments- Use *args and **kwargs  for flexibility

```python
def add_all(*args):
    return sum(args)

add_all(1,2,3)
```

```python
def show_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
show_info(name="Alice", age=25)
```

```python
def student_info(**student):
    return f"My name is {student["name"]} \nI am {student["age"]} years old and\n I am in grade {student["grade"]} "

student_info(name="Joy", age= 14, grade= 4)
```

```python
def process_order(customer_name, table_number, *dishes, drink="Water",dessert="None",**special):
    return f"Order for {customer_name} at table {table_number}\nMain Dishes: {dishes}\nDrink: {drink}\nDessert: {dessert} \nSpecial Instructions:\n- spicy_level: {special['spicy_level']} "

print(process_order("Alice",5, "Pizza","Pasta",
             drink="Coke",
              spicy_level="Medium",
              allergies= "Nuts"
             ))
    
```

```python
def order_summary(name,*items,delivery="pickup", **details):
    return f"{name}\n{items} \n{delivery}\n{details['location']\n{details['phone']}}"
print(order_summary(
    "Abigael",
    "Bread","Milk","Sugar",
    delivery="Home",
    location="Nairobi",
    phone="0725875650"
))
```

```python
def add_numbers(number,*nums,show_steps="True",**info):
    total = number
    for n in nums:
        if show_steps =="True":
            print("Adding", n)
        total += 1 
    print(info)
    return total
    
add_numbers(
    20,20,20,
    show_steps="False",
    user="Aisha",
    role ="Student"
)
```
